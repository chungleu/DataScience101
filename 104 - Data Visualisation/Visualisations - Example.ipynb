# Imports

```python
import matplotlib
import matplotlib.pyplot as plt
import numpy as np
```

# Simple Chart

```python
t = np.arange(0, 2, 0.01)
s = 1 + np.sin(2 * np.pi * t)

fig, ax = plt.subplots()

ax.plot(t, s)
ax.set(xlabel='time (s)', ylabel='voltage (mV)',
      title='Nice and simple')
ax.grid()

plt.show()

fig.savefig('simple wavey.png')
```
