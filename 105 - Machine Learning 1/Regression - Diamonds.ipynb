## Linear Regression Model

```python
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error
```

```python
X_train, X_test, y_train, y_test = train_test_split(features, target, random_state=42)
```

```python
X_train.shape
```

```python
len(X_train)/len(features)
```

```python
regr = LinearRegression()
```

```python
model = regr.fit(X_train, y_train)
```

```python
features.columns
```

```python
model.predict([[1.5, 45, 45, 4, 4, 4]])
```

```python
predicted = model.predict(X_train)
predicted
```

```python
model.score(X_train, y_train)
```

```python
model.score(X_test, y_test)
```

```python
print('Training data correlation:', model.score(X_train, y_train))
print('Test data correlation:', model.score(X_test, y_test))
```

```python
import numpy as np
```

```python
plt.figure(figsize=(15,8))

plt.scatter(x=y_train, y=predicted, alpha=0.1, color='purple')

plt.plot(np.unique(y_train), np.poly1d(np.polyfit(y_train, predicted, 1))(np.unique(y_train)), color='royalblue')

plt.xlabel('Actual prices')
plt.ylabel('Predicted prices')

plt.show()
```

```python
mse = mean_squared_error(y_train, predicted)
```

```python
import math
math.sqrt(mse)
```

### Ridge Regression
A way to 'regularise' our model. Adds a penalty as the model complexity increases <br>
Minimises the impact of overfitting by (applying weighted) penalties parameters in complex models

```python
from sklearn.linear_model import Ridge
```

```python
ridge_regr = Ridge(alpha=1.0)
```

```python
ridge_model = ridge_regr.fit(X_train, y_train)
```

```python
ridge_predicted = ridge_model.predict(X_train)
ridge_predicted
```

```python
plt.figure(figsize=(15,8))

plt.scatter(x=y_train, y=ridge_predicted, alpha=0.1, color='purple')

plt.plot(np.unique(y_train), np.poly1d(np.polyfit(y_train, ridge_predicted, 1))(np.unique(y_train)), color='royalblue')

plt.xlabel('Actual prices')
plt.ylabel('Ridge Predicted prices')

plt.show()
```

```python
print('Linear Regression Score on Training Data:', model.score(X_train, y_train))
print('Linear Regression Score on Test Data:', model.score(X_train, y_train))
print('Ridge Regression Score on Training Data:', ridge_model.score(X_train, y_train))
print('Ridge Regression Score on Test Data:', ridge_model.score(X_train, y_train))
```

### Polynomial

```python
from sklearn.preprocessing import PolynomialFeatures
```

```python
degrees = 2
```

```python
X_poly = PolynomialFeatures(degree=degrees).fit_transform(X_train)
```

```python
X_poly_test = PolynomialFeatures(degree=degrees).fit_transform(X_test)
```

```python
X_poly_test.shape
```

```python
y_test.shape
```

```python
regr2 = LinearRegression()
regr2.fit(X_poly, y_train)
regr2.score(X_poly, y_train)
```

```python
print('Linear Regression Score on Training Data:', model.score(X_train, y_train))
print('Linear Regression Score on Test Data:', model.score(X_test, y_test))
print('Ridge Regression Score on Training Data:', ridge_model.score(X_train, y_train))
print('Ridge Regression Score on Test Data:', ridge_model.score(X_test, y_test))
print('Polynomial Regression Score on Training Data:', regr2.score(X_poly, y_train))
print('Polynomial Regression Score on Test Data:', regr2.score(X_poly_test, y_test))
```

### Polynomial - Make Pipeline

```python
from sklearn.pipeline import make_pipeline
```

```python
poly_regr = make_pipeline(PolynomialFeatures(degrees), LinearRegression())
```

```python
poly_regr.fit(X_train, y_train)
```

```python
print(poly_regr.score(X_train, y_train))
print(poly_regr.score(X_test, y_test))
```

```python
poly_predicted = poly_regr.predict(X_test)
poly_predicted
```
